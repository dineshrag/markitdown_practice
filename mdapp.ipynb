<h1>3. Get Transcript from Youtube video and Summarize using LLAMA3.2 Local model</h1>

```python
result_transcript = md.convert_url("https://www.youtube.com/watch?v=HdPzOWlLrbE")
print(result_transcript.text_content)
```

```python
import re
transcript = re.search(r"### Transcript\s*(.*?)\s*(?=(\n###|\Z))", result_transcript.text_content, re.S)
transcript_text = ""
if transcript:
    transcript_text = transcript.group(1).strip()
else:
    print("Transcript not found.")
print(transcript_text)
```

```python
message = {
    'role': 'user',
    'description': 'You are an AI assistant helping users to summarize the provided content. Please provide summary in 5 bullet points.',
    'content': transcript_text
}
```

```python
import ollama
# Use the ollama.chat function to send the transcript and retrieve the summary
response = ollama.chat(
    model="llama3.2",
    messages=[message]
)
```

```python
print(response.message.content)
```

<h1>4. Description for the image using llava</h1>

```python
# Path to your image
image_path = 'data/quote.jpg'

# Prepare the message to send to the LLaVA model
message = {
    'role': 'user',
    'content': 'Describe this image:',
    'images': [image_path]
}
```

```python
# Use the ollama.chat function to send the image and retrieve the description
response = ollama.chat(
    model="llava",
    messages=[message]
)
```

```python
# Print the model's description of the image
print(response.message.content)
```
